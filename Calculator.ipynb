```python
sales = float(input("Enter sales value ($): "))
tax_rate = float(input("Enter Tax rate (%): "))

cogs = 0.4 * sales
gross_profit = sales - cogs
taxes = gross_profit * (tax_rate / 100)
net = gross_profit - taxes

print(f"Sales: {sales}$")
print(f"Cost of goods sold (COGS): {cogs}$")
print(f"Gross profit: {gross_profit}$")
print(f"Taxes: {taxes}$")
print(f"Net income: {net}$")
```
