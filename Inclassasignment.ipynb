```python
sales = float(input("enter sales ($):"))
```

```python
rate = float(input("enter Tax Rate (%)"))
```

```python
cogs = sales *0.4
gross_profit = sales - cogs
taxes = rate * gross_profit
net = gross_profit - taxes
```

```python
print(f"sales: {sales}$")
print(f"Costs of Goods Sold: {cogs}$")
print(f"Gross Profit: {gross_profit}$")
print(f"Taxes: {taxes}$")
print(f"Net Income: {net}$")
```
